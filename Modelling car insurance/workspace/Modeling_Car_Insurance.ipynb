```python
insurance_cat = insurance.copy()

for col in ['age', 'gender', 'driving_experience', 'education','income','vehicle_ownership','vehicle_year','married','children','vehicle_type','outcome']:
    insurance_cat[col] = insurance_cat[col].astype('category')
    
insurance_cat.dtypes
```

### Preparing For Modelling

```python
model_list = []

features = insurance_cat.drop(columns = ['outcome','id']).columns
features
```

```python
model_dict = {}

for feature in features:
    featurette = logit(f'outcome ~ {feature}', data = insurance).fit()
    model_list.append(featurette)
    # name = f'mdl_outcome_{feature}'
    model_dict[f'mdl_outcome_{feature}'] = featurette

print(lines)
print(model_list,'\n\n\n')
print(model_dict)
```

```python
for model in model_list:
    print(model.params,'\n\n')
```

```python
accuracies = []
accuracy_dict = {}

for feature in range(0, len(model_list)):
    # print(model.params,'\n\n')
    conf_matrix = model_list[feature].pred_table()
    # True negatives
    tn = conf_matrix[0,0]
    # True positives
    tp = conf_matrix[1,1]
    # False negatives
    fn = conf_matrix[1,0]
    # False positives
    fp = conf_matrix[0,1]
    # Compute accuracy
    acc = (tn + tp) / (tn + fn + fp + tp)
    accuracies.append(acc)
    accuracy_dict[feature] = acc
    
```

```python
print(accuracies)
```

```python
print(accuracy_dict)
```

```python
best = accuracies.index(max(accuracies))
# print(type(features))
features[2]
```

```python
# Find the feature with the largest accuracy
best_feature = features[accuracies.index(max(accuracies))]
best_feature
```
